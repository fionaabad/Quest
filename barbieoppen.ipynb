## Barbie Ordering and filtering

### Domestic

```python
drop_columns = ['Rank', 'Theaters', 'Per Theater']
df_dom_barbie1 = df_domestic_box_barbie.drop(drop_columns, axis=1)
df_dom_barbie2 = df_dom_barbie1.drop(0)

df_dom_barbie3= df_dom_barbie2.drop(df_dom_barbie2.columns[2], axis=1)
df_dom_barbie = df_dom_barbie3.drop(df_dom_barbie3.columns[2], axis=1)
df_dom_barbie
```

#### Ingresos totales

```python
# we have to replace the $ sing and turn it into a float
df_dom_barbie['Gross'] = df_dom_barbie['Gross'].str.replace('$', '').str.replace(',', '')
df_dom_barbie['Gross'] = df_dom_barbie['Gross'].astype(float)

total_gross = df_dom_barbie['Gross'].sum()
formatted_total_gross = '{:,.0f}'.format(total_gross)

print("Ingresos totales: $", formatted_total_gross)
```

#### Dia con mayor ingresos

```python
date_with_highest_gross = df_dom_barbie['Date'].iloc[df_dom_barbie['Gross'].idxmax()]
print("Fecha con la mayor recaudación: ", date_with_highest_gross)
```

#### Observamos los cambios de un dia para otro

```python
# We use the pct_change(): is a function that calculates the percentage change between adjacent elements in a series or column of a DataFrame.
df_dom_barbie['Growth'] = (df_dom_barbie['Gross'].pct_change() * 100).round(2)
df_dom_barbie
```

#### Vamos acomulando los ingresos

```python
# We use cumsum(): calculate the cumulative sum of values in a series or column of a DataFrame. 
# It returns a new series or column with the accumulated or running sum at each position.
df_dom_barbie['Total Gross'] = df_dom_barbie['Gross'].cumsum()
df_dom_barbie
```

### International

```python
drop_columns= ['Opening Weekend','Opening Weekend Screens', 'Maximum Screens', 'Theatrical Engagements', 'Report Date' ]
df_int_barbie = df_international_box_barbie.drop(drop_columns, axis=1)
```

```python
# convertimos la columna de Total Box Office en un float
df_int_barbie['Total Box Office'] = df_int_barbie['Total Box Office'].astype(str)
df_int_barbie['Total Box Office'] = df_int_barbie['Total Box Office'].str.replace('$', '').str.replace(',', '').astype(float)
```

#### Ingresos Totales

```python
total_gross2 = df_int_barbie['Total Box Office'].sum()
formatted_total_gross2 = '{:,.0f}'.format(total_gross2)

print("Ingresos totales: $", formatted_total_gross2)
```

#### Group by bigger box office

```python
int_barbie = df_int_barbie.sort_values('Total Box Office', ascending=False)
int_barbie2 = int_barbie.reset_index()
barbie_int = int_barbie2.drop(26).drop(25).drop(24).drop(23).drop(22).drop(21).drop(20)
```
